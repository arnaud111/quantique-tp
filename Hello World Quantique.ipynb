```python
import numpy as np
```

```python
from qiskit import *
```

```python
# On crée un circuit quantique
```

```python
circ = QuantumCircuit(3)
```

```python
# Affichons le cicruit
```

```python
circ.draw()
```

```python
circ.draw('mpl')
```

```python
# mettons un qubit en état de superposition
```

```python
circ.h(0)
```

```python
# h = porte de Hadamard = superposition d'un qubit
```

```python
circ.draw('mpl')
```

```python
# Intriquons  qubits
```

```python
circ.cx(0,1)
```

```python
# cx = CNOT = Control Not = intricaton de deux qubits
```

```python
circ.draw()
```

```python
# Intriquons les qubits 0 et 2
```

```python
circ.cx(0,2)
```

```python
circ.draw('mpl')
```

```python
# Utilisons un simulateur quantique
```

```python
# Importe Aer = le module des simulateurs
```

```python
from qiskit import Aer
```

```python
# On crée notre moteur de calcul
```

```python
backend = Aer.get_backend('statevector_simulator')
```

```python
# On lance le calcul quantique
```

```python
job = backend.run(circ)
```

```python
# Affichons le résultat
```

```python
result = job.result()
```

```python
# On a simplement des vecteurs d'état
```

```python
outputstate = result.get_statevector(circ,decimals=3)
```

```python
print(outputstate)
```

```python
# Affichons un histogramme
```

```python
from qiskit.visualization import plot_state_city
```

```python
plot_state_city(outputstate)
```
